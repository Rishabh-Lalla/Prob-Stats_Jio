### Example (Discrete random variable and PMF)

An unfair coin for which $P(H)=p$, where 0<p<1. I toss the coin repeatedly until I observe a heads for the first time. Let $Y$ be the total number of coin tosses.
- Suggest a distribution for $Y$.
- Verify it is a PMF.
- If $p= \frac{1}{2}$, find $P(2≤Y<5)$.

**Solution:**

First, we note that the random variable $Y$ can potentially take any positive integer, so we have $R_Y=N=\{1,2,3,...\}$. To find the distribution of $Y$, we need to find $P_{Y}(k)=P(Y=k)$ for $k=1,2,3,....$ We have

$P_{Y}(1)=P(Y=1)=P(H)=p$,

$P_{Y}(2)=P(Y=2)=P(TH)=(1−p)p$,

$P_{Y}(3)=P(Y=3)=P(TTH)=(1−p)^2p$,
.                  

.

.

$P_{Y}(k)=P(Y=k)=P(TT...TH)=(1−p)^{(k−1)}p$.


Thus, we can write the PMF of Y in the following way
$$
P_{Y}(y)= \begin{cases}(1-p)^{y-1} p & \text { for } y=1,2,3, \ldots \\ 0 & \text { otherwise }\end{cases}
$$



To check that $\sum_{y \in R_{Y}} P_{Y}(y)=1$, we have
$$
\begin{aligned}
\sum_{y \in R_{Y}} P_{Y}(y) &=\sum_{k=1}^{\infty}(1-p)^{k-1} p \\
&=p \sum_{j=0}^{\infty}(1-p)^{j} \\
&=p \frac{1}{1-(1-p)} \\
&=1
\end{aligned}
$$


If $p=\frac{1}{2}$, to find $\mathrm{P}(2 \leq Y<5)$, we can write
$$
\begin{aligned}
\mathrm{P}(2 \leq Y<5) &=\sum_{k=2}^{4} P_{Y}(k) \\
&=\sum_{k=2}^{4}(1-p)^{k-1} p \\
&=\frac{1}{2}\left(\frac{1}{2}+\frac{1}{4}+\frac{1}{8}\right) \\
&=\frac{7}{16} .
\end{aligned}
$$



### Example: (Properties of Expectations)

Sometime it may be difficult to find the probability mass function of a random variable directly, but it may still be quite easy to find the expectation of the random variable. As we have seen in class, the expectation is a very important summary characteristic of a random variable.

There is a dinner party where $n$ men check their hats. The hats are mixed up during dinner, so
that afterward each man receives a random hat. In particular, each man gets his own hat with
probability $\frac{1}{n}$. What is the expected number of men who get their own hat?


**Solution**

Define,
$G:$ be the number of men that get their own hat, we want to find the expectation of $G$.
we don’t know enough about the distribution of $G$.

Let,
$G_i:$ Event that the $i^{th}$  man gets his own hat.

That is, $G_i = 1$ if he gets his own hat, and $G_i = 0$ otherwise. The number of men that get their own hat is the sum of these indicators: $G= G_1+G_2+...+G_n$

Now since $G_i$ is an indicator, we know $1/n = P (G_i = 1) = E [G_i]$

$$
\begin{aligned}
\mathrm{E}[G] &=\mathrm{E}\left[G_{1}+G_{2}+\cdots+G_{n}\right] \\
&=\mathrm{E}\left[G_{1}\right]+\mathrm{E}\left[G_{2}\right]+\cdots+\mathrm{E}\left[G_{n}\right] \\
&=\frac{1}{n}+\frac{1}{n}+\cdots+\frac{1}{n}=n\left(\frac{1}{n}\right)=1
\end{aligned}
$$


### Example:  (Variance in games)

**Game A:** We win  2 dollars with probability 2/3 and lose 1 dollar with probability 1/3.


**Game B:** We win 1,002  dollars with probability 2/3 and lose 2,001 dollars with probability 1/3.

Which game is better financially?

**Solution:**

Note that probability of winning for each game is same.

What about the expected return for each game?

Let random variables $A$ and $B$ be the payoffs for the two games.



$$
\begin{array}{r}
\mathrm{E}[A]=2 \cdot \frac{2}{3}+(-1) \cdot \frac{1}{3}=1 \\
\mathrm{E}[B]=1002 \cdot \frac{2}{3}+(-2001) \cdot \frac{1}{3}=1
\end{array}
$$


The expected payoff is the same for both games, but they are obviously very different! This difference is not apparent in their expected value, but is captured by variance. We can compute the $\operatorname{Var}[A]$ by working "from the inside out" as follows:

$$
\begin{aligned}
A-\mathrm{E}[A] &=\left\{\begin{array}{cc}
1 & \text { with probability } \frac{2}{3} \\
-2 & \text { with probability } \frac{1}{3}
\end{array}\right.\\
(A-\mathrm{E}[A])^{2} &= \begin{cases}1 & \text { with probability } \frac{2}{3} \\
4 & \text { with probability } \frac{1}{3}\end{cases} \\
\mathrm{E}\left[(A-\mathrm{E}[A])^{2}\right] &=1 \cdot \frac{2}{3}+4 \cdot \frac{1}{3} \\
\operatorname{Var}[A] &=2 .
\end{aligned}
$$


Similarly, we have for $\operatorname{Var}[B]$ :


$$
\begin{aligned}
B-\mathrm{E}[B] &=\left\{\begin{array}{cc}
1001 & \text { with probability } \frac{2}{3} \\
-2002 & \text { with probability } \frac{1}{3}
\end{array}\right.\\
(B-\mathrm{E}[B])^{2} &=\left\{\begin{array}{cc}
1,002,001 & \text { with probability } \frac{2}{3} \\
4,008,004 & \text { with probability } \frac{1}{3}
\end{array}\right.\\
\mathrm{E}\left[(B-\mathrm{E}[B])^{2}\right] &=1,002,001 \cdot \frac{2}{3}+4,008,004 \cdot \frac{1}{3} \\
\operatorname{Var}[B] &=2,004,002 .
\end{aligned}
$$

High variance is often associated with high risk. For example, in ten rounds of Game A, we expect
to make 10 dollars, but could conceivably lose 10 dollars instead. On the other hand, in ten rounds of game B,
we also expect to make 10 dollars, but could actually lose more than 20,000 dollars!

### Example (Identify the discrete probability distribution for this problem and simple probabilities)

An urn contains $N$ white and $M$ black balls. Balls are randomly selected, one at a
time, until a black one is obtained. If we assume that each ball selected is replaced
before the next one is drawn, what is the probability that
- exactly n draws are needed?
- at least k draws are needed?

**Solution:**

$X:$ denote the number of draws needed to select a black ball

Also Define, $p$= $P$(black ball is obtained)= $\frac{M}{(M+N)}$



$$
P[X=n]=\left(\frac{N}{M+N}\right)^{n-1} \frac{M}{M+N}=\frac{M N^{n-1}}{(M+N)^{n}}
$$

This is the PMF of a Geometric distribution with $p=\frac{M}{M+N}$.


$$
\begin{aligned}
P[X \geq k] &=\frac{M}{M+N} \sum_{n=k}^{\infty}\left(\frac{N}{M+N}\right)^{n-1} \\
&=\left(\frac{M}{M+N}\right)\left(\frac{N}{M+N}\right)^{k-1} /\left[1-\frac{N}{M+N}\right] \\
&=\left(\frac{N}{M+N}\right)^{k-1}
\end{aligned}
$$

### Example: Random Walk

A drunkard stands at some integer on a number line. At each step he moves either left or right to the adjacent number. He chooses the left or right with equal probability. Let $S_n$ be his position after $n$ steps.At each time point, we'll keep track of our position and then plot our trajectory.


```python
# For scientific computing
import numpy as np

# For plotting
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('notebook')
%matplotlib inline

# Define our step probability and number of steps.
step_prob = 0.5  # Can step left or right equally.
n_steps = 1000   # Essentially time.

# Set up a vector to store our positions.
position = np.zeros(n_steps)  # Full of zeros.

# Loop through each time step.
for i in range(1, n_steps):
    # Flip a coin.
    flip = np.random.rand()

    # Figure out which way we should step.
    if flip < step_prob:
        step = -1  # To the 'left'.
    else:
        step = 1  # to the 'right'.

    # Update our position based off of where we were in the last time point.
    position[i] = position[i-1] + step


# Make a vector of time points.
steps = np.arange(0, n_steps, 1)  # Arange from 0 to n_steps taking intervals of 1.

# Plot it!
plt.plot(steps, position)
plt.xlabel('number of steps')
plt.ylabel('position')
```

```python
# Perform the random walk 1000 times.
n_simulations = 1000

# Make a new position vector. This will include all simulations.
position = np.zeros((n_simulations, n_steps))

# Redefine our step probability just to be clear.
step_prob = 0.5

# Loop through each simulation.
for i in range(n_simulations):
    # Loop through each step.
    for j in range(1, n_steps):
        # Flip a coin.
        flip = np.random.rand()

        # Figure out how to step.
        if flip < step_prob:
            step = -1
        else:
            step = 1

        # Update our position.
        position[i, j] = position[i, j-1] + step

# Plot all of the trajectories together.
for i in range(n_simulations):
    # Remembering that `position` is just a two-dimensional matrix that is
    # n_simulations by n_steps, we can get each step for a given simulation
    # by indexing as position[i, :].
    plt.plot(steps, position[i, :], linewidth=1, alpha=0.5)

# Add axis labels.
plt.xlabel('number of steps')
plt.ylabel('position')
```

# Now our objective is to find the distribution of of $S_n$. Let's choose n=50, and draw a histogram for $S_{50}$

```python
# Make a histogram of the positions. To look at t=200, we have to index at
# 199 because indexing starts at 0  in Python. We'll also normalize the
# histogram (normed=True) so we can get a measure of probability.
plt.hist(position[:, 49], bins=20, normed=True)
plt.xlabel('position')
plt.ylabel('probability')

# Set the xlimits to cover the entire range.
plt.xlim([-50, 50])
```

### Example: Random Graph Model

$\color{red}{\text{What is a random graph?}}$

$\color{red}{\text{How do we create a random unweighted, undirected graph on $n$ nodes?}}$

- **Idea 1:**  From the set of all graphs of $n$ nodes, pick one uniformly at random.
- **Idea 2:**  : Specify a probability $0 ≤ p ≤ 1$. For every pair of nodes, add an edge between the nodes with probability $p$.


**Idea 1:**



- Expected degree of a node is : $\frac{n-1}{2}$
- Average numbers of edges for the graph $\frac{n(n-1)}{4}$


**Idea 2 ( Erdos-Renyi Model)**

$G(n,p):$ an undirected, unweighted graph (family) with $n$ node.

##### Question: How to generate a graph in $G(n,p)$?


- For each pair $(u, v)$ of $\left(\begin{array}{l}n \\ 2\end{array}\right)$ node pairs, connect $u$ and $v$ by an edge with probability $p$.
- How do you "do something" with probability $p$ ?
- Generate a random number $x$ between 0 and 1 under the uniform distribution. If $x \leq p$, then "do something", else "do the other thing"  


Average no of edges does this graph have:
$\frac{n(n − 1)p}{2}$

##### Question: Degree Distribution of $G(n,p)$

Now our objective is to calculate the the probability of a vertex will have a certain degree $k$.
- Connect (probability p) $v$ to $k$ neighbours out of $n-1$ nodes and not connect (probability $1-p$ ) to the rest.
- Probability that $v$ has degree $k$ follows the bino
mial distribution
$$
\operatorname{Pr}(d(v)=k)=\left(\begin{array}{c}
n-1 \\
k
\end{array}\right) p^{k}(1-p)^{n-k-1}
$$

It is clear that vertex has a binomial distribution with parameter $(n-1)$ and $p$.
Therefore, expected degree of a node/vertex is : $(n-1)p$
