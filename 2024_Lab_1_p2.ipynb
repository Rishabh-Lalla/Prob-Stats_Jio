### Example 2: Construction of Sample Space
You get a stick and break it randomly into three pieces. What is the probability that you can make a triangle using the three pieces? You can assume the break points are chosen completely at random, i.e. if the length of the original stick is 1 unit, and $x,y,z$ are the lengths of the three pieces, then $(x,y,z)$ are uniformly chosen from the set
$$\{(x,y,z)∈\mathbb{R}^3|x+y+z=1,x,y,z≥0\}$$

#### Solution:
Note that this is again a problem on a continuous probability space.

First, we need to identify the sample space S. In this case the sample space is going to be a two-dimensional set.

Second, we need to identify the set A that contains the favorable outcomes (the set of (x,y,z) in S that form a triangle).

Finally, since the space is uniform, we will divide area of set A by the area of S to obtain P(A).

Note that in $\mathbb{R^3}$, x+y+z=1 represents a plane that goes through the points $(1,0,0),(0,1,0),(0,0,1)$. To find the sample space S, note that $S={(x,y,z)∈\mathbb{R}^3|x+y+z=1,x,y,z≥0}$.

To find the set A, note that we need $(x,y,z)$ to satisfy the triangle inequality
$x+y>z$,
$y+z>x$,
$x+z>y$.

<!--  -->

Note that since $x+y+z=1$, we can equivalently write the three equations as
$x<\frac{1}{2}$, $y<\frac{1}{2}$, $z<\frac{1}{2}$.

Therfore $\mathbb{P}(A)=\frac{\text{Area of A}}{\text{Area of S}}$

### Example 3: Urn Scheme

Suppose an urn has $r$ red balls and $B$ black balls. A ball is drawn and its color noted. Then it together with $c >0$ balls of the same color as the drawn ball are added to the urn. The procedure is repeated $n-1$ times so that the total number of drawings made from the urn is $n$.

Define,

$R_j, 1 \leq j \leq n : $ event that the $j^{th}$ ball drawn is red.

$B_j, 1 \leq j \leq n : $ event that the $j^{th}$ ball drawn is black.



$$
P\left(R_{1} \cap R_{2}\right)=P\left(R_{1}\right) P\left(R_{2} \mid R_{1}\right)
$$
Now
$$
P\left(R_{1}\right)=\frac{r}{b+r}, \quad P\left(R_{2} \mid R_{1}\right)=\frac{r+c}{b+r+c},
$$
and thus


$$
P\left(R_{1} \cap R_{2}\right)=\left(\frac{r}{b+r}\right)\left(\frac{r+c}{b+r+c}\right) .
$$
Similarly


$$
P\left(B_{1} \cap R_{2}\right)=\left(\frac{b}{b+r}\right)\left(\frac{r}{b+r+c}\right)
$$
and thus


$$
\begin{aligned}
P\left(R_{2}\right) &=P\left(R_{1} \cap R_{2}\right)+P\left(B_{1} \cap R_{2}\right) \\
&=\left(\frac{r}{b+r}\right)\left(\frac{r+c}{b+r+c}\right)+\left(\frac{b}{b+r}\right)\left(\frac{r}{b+r+c}\right) \\
&=\frac{r}{b+r} .
\end{aligned}
$$

Consequently, $P(R_2)= P(R_1)$.

Now show that probability of selecting a red ball at any trial $n$ is $\frac{r}{b+r}$. This is known as Polya's urn scheme.

### Example 4: Flights and Rain

Mumbai airport hires you to estimate how the punctuality of flight arrivals is affected by the weather.

You begin by defining a probability space, with sample space all the possible outcomes:

$$
\Omega=\{\text{late and rain; late and no rain; on time and rain; on time and no rain}\}
$$

You examine data from previous flights and determine the following probabilities for each simple event

$$
P(\text{late and rain})=\frac{3}{20}
$$

$$
P(\text{late and no rain})= \frac{2}{20}
$$

$$
P(\text{on time and rain})=\frac{1}{20}
$$

$$
P(\text{on time and no rain})=\frac{14}{20}
$$

The airport is interested in the probability of a flight being late if it rains, that is in the **conditional probability**
$
P(\text{late}\mid \text{rain})
$
which we can obtain by computing

$$
P(\text{late}\mid \text{rain})= \frac{P(\text{late and rain})}{P(\text{rain})}
$$

The probability at the numerator is available from historical data. We compute the probability at the denominator as the union of the probabilities of the disjoint events  $\text{on time and rain}$ and $\text{late and rain}$:

$$
P(\text{rain})=P(\text{on time and rain})+P(\text{late and rain})= \frac{1}{20}+ \frac{3}{20}= \frac{4}{20}
$$

Thus,

$$
P(\text{late}\mid \text{rain})= \frac{P(\text{late and rain})}{P(\text{rain})}=\frac{3/20}{4/20}= \frac{3}{4}
$$

That is there is a $75\%$ chance that a flight is late when it rains!

Note that we can also compute the probability of a late flight if it doesn't rain:

$$
P(\text{late}\mid \text{no rain}=\frac{P(\text{late and no rain})}{P(\text{ no rain})}
$$

Since $P(\text{ no rain})=1-P((\text{rain})=1- \frac{4}{20}= \frac{16}{20}$, then

$$
P(\text{late}\mid \text{no rain}=\frac{P(\text{late and no rain})}{P(\text{ no rain})}= \frac{2/20}{16/20}= \frac{1}{8}=0.125
$$

So, there is only a 12.5\% chance that the flight is late if it doesn't rain.

Now, your aunt arrives at the Mumbai airport tomorrow and you would like to know how likely it is for her flight to be on time. This means that we would like to compute the $P(\text{late})$ for a general traveler (your aunt). We can do this by computing

$$
P(\text{late})=P(\text{late and no rain})+P(\text{late and rain})=\frac{2}{20}+\frac{3}{20}= \frac{5}{20}=0.25
$$

We could have also used the  **law of total probability**

$$
P (\text{late}) = P (\text{late}\mid \text{rain}) P (\text{rain}) + P (\text{late}\mid \text{no rain}) P (\text{no rain})= 0.75 \times 0.2 + 0.125 \times 0.8 = 0.25.
$$

Your aunt tells you that she arrived late at the aiport, but forgot to tell you if it was raining or not. You want to determine the probability that it was raining, based on the fact that your aunt was late.

We can do this by using the **Bayes theorem**:

$$
P(\text{rain}|\text{late})=\frac{P(\text{late and rain})}{P(\text{late})}=\frac{\frac{3}{20}}{\frac{1}{4}}= \frac{3}{5}=0.6
$$

Now suppose that you want to book a taxi for your aunt. You know that the probability of finding a taxi depends on the weather at the airport, more precisely you have had the opportunity to compute that

$$
P(\text{taxi}\mid \text{rain})=0.1 \text{ and } P(\text{taxi}\mid \text{no rain})=0.6
$$

## Simulations: A first example of Monte Carlo approximation

It is often possible to use computers to simulate and approximate the theoretical results.

Computer simulation essentially does in actual code what one does mathematically with probability calculations.


#### Example:  Rolling Dice

Suppose we are interested in answering the following question:

*If we roll three dice, what is the probability that their total is 8?*

We could count all the possibilities, or we could get an approximate answer via simulation.

To do that, we can write a small **Python program** like the one below.

Note that I am leaving the number of dice that are thrown as a quantity to be fixed any time I need.

So the following snippet of code also work is I am interested in the event

$$
P(\{\text{obtaining a total of $k$ when laungching \texttt{num_dice} dices}\})
$$

where you can fix $k$ and \texttt{num_dice} as needed. In the example above, $k$=8 and \texttt{num_dice}$=3$.



```python
# importing numpy
import numpy as np
import pandas as pd
import random
np.random.seed(1234)

def roll_dice(num_dice):
    """Return a list of integers with length `num_dice`.

    Each integer in the returned list is a random number between
    1 and 6, inclusive.
    """
    roll_results = []
    for _ in range(num_dice):
        roll = random.randint(1, 6)
        roll_results.append(roll)
    return roll_results

def sum_simulation(num_dice, k, nreps):
    sums = np.zeros(nreps)
    for rep in range(nreps):
         sums[rep]=sum(roll_dice(num_dice))
    return np.count_nonzero(sums == k)/nreps
```

We can check the simulation results for different values of the simulation repetitions (length).

We can start with a question we know already the answer of (we did it in class). The probability of having a sum of 7 when rolling two dice. We know that probability is 6/36=0.1666. We can see how the accuracy changes with increasing the number of simulations in this case

```python
simulation_result=sum_simulation(2, 7, 30) #10 repetitions
print(simulation_result)
```

```python
simulation_result=sum_simulation(2, 7, 10) #10 repetitions
print(simulation_result)

simulation_result=sum_simulation(2, 7, 100) #100 repetitions
print(simulation_result)

simulation_result=sum_simulation(2, 7, 1000) #1000 repetitions
print(simulation_result)

simulation_result=sum_simulation(2, 7, 100000) #100,000 repetitions
print(simulation_result)
```

Now, let's change the parameters of the simulation for the question we have asked above: the probability of having a sum of 8 when rolling three dice. This can be obtained by changing the original code as follows:

```python
simulation_result=sum_simulation(3, 8, 100000) #100,000 repetitions
print(simulation_result)
```

Using simulations may be much simpler than doing computations.
However, in order to compute probability of complex events,
we may need to know the formulas required to compose the results from multiple simulations!

For example, suppose three fair dice are rolled. We wish to find the approximate probability that the  first die
shows fewer than 3 dots, given that the total sum for the three dice is more than 8, using simulation.

We need to compute:

$$
P(\text{first die} < 3 |\text{sum} > 8)=\frac{P(\{\text{first die} < 3\}\cap \{\text{sum} > 8\})}{P(\text{sum} > 8)}
$$

We can modify the previous functions in the following way to find what we need

```python
def sum_simulation_2(num_dice, k, nreps):
    sums = np.zeros(nreps)
    first_die=np.zeros(nreps)
    for rep in range(nreps):
     roll_result=roll_dice(num_dice)
     first_die[rep]=roll_result[0] #obtain first element
     sums[rep]=sum(roll_result) #get the sum
    # obtain probabilities of interest
    joint_probability=np.count_nonzero((first_die < 3) & (sums >= 8))/nreps
    marg_probability_sum=np.count_nonzero(sums >= 8)/nreps
    return joint_probability, marg_probability_sum


joint, marg = sum_simulation_2(3, 8, 100000)
print(joint)
print(marg)
```

We can then use these probabilities to compute the probability we are interested in:

$$
P(\text{first die} < 3 |\text{sum} > 8)=\frac{P(\{\text{first die} < 3\}\cap \{\text{sum} > 8\})}{P(\text{sum} > 8)}=\frac{0.21606}{0.83805}=0.2578
$$

#### Appendix Example 1: The Importance of Scoring First in Soccer

(Taken from Severini (2020), "Analytic Methods in Sports", CRC press)

We consider data from soccer games played in the English Premier League (EPL) in the 2010-2011 through 2012-
2013 seasons. The data analyzed here are available on SoccerSTATS.com.

Based on these data, the probability that the home team wins the game is
0.453. Note that this is an average value for the entire league; we can think of
it as applying to a game played between two "randomly chosen" EPL teams.

We can write this result as

$$
P \text { (Home team wins) }=0.453 .
$$

Now, suppose that we include the additional information that the home
team scores first. This additional information changes the probability that the
home team wins the game. For the EPL, the conditional probability that the
home team wins given that the home team scores first is \(0.718\). We write this
as


$$
\qquad P (\text{ Home team wins }\mid \text{ Home team scores first})=0.718
$$


Therefore, the fact that the home team scores first as an important effect on the probability that the home team wins the game, changing it from the unconditional probability of 0.453 to 0.718.

From the same data, we can compute (well, to be more precise, estimate) the probability that the home team wins as $0.178$,

$$
P(\text{ Home team wins}\mid \text{Visiting team scores first})=0.178$
$$


Again, the fact that the visiting team scores first is important information
that greatly affects the probability that the home team wins. There is the
possibility that neither team scores first - that is, there are no goals scored - in this case, the game is a draw so that

$$
P(\text { Home team wins } \mid \text { Neither team scores first) })=0.076\text {. }
$$

These three conditional probabilities are related. The probability that the
home team scores first is $0.534$, the probability that the visiting team scores
first is $0.390$, and the probability that neither team scores first is $0.076$.

The law of total probability discussed can be applied to these three conditioning events:

<span style="color:blue"><br/>
$$\begin{aligned}
P(\text{Home team wins})&=P(\text{ Home scores first}) \times P(\text{Home team wins} \mid \text{Home scores first})\\
&+P(\text{Visitor scores first}) \times P(\text{Home team wins}\mid \text{Visitor scores first})\\
&+P(\text{Neither scores first}) \times P(\text{Home team wins}\mid\text{Neither scores first}).
\end{aligned}$$
</span><br/>

Using the probability values based on the EPL data, this expression becomes

<span style="color:blue"><br/>
$$
P(\text{Home team wins})=(0.534)(0.718)+(0.390)(0.178)+(0.076)(0)=0.453
$$
</span><br/>

What is the probability the the home team wins if they hare scored first? We can use Bayes theorem:

<span style="color:blue"><br/>
$$
\begin{aligned}
P(\text { Home team scores first | Home team wins })&=\frac{P(\text{Home team wins}|\text{Home team scores first})\, P(\text{Home team scores first})}{P(\text{Home team scores first})}\\&=\frac{(0.718) \times 0.534}{0.453}=0.846
\end{aligned}
$$
</span><br/>

#### Food for thought: computing winning probabilities as the game unfolds

The previous analysis shows how the probability of an event, such as
the home team winning a soccer game, can be updated to reflect additional
information, such as the fact that the home team scored the first goal of the
game, using conditional probabilities.

An extension of this idea is the concept of "win probabilities," which give the probability of a team winning a game
as a function of the current status of the game.

Although the basic approach can be applied to any sport, here we consider it in the context of football.

Consider an NFL game; because the win probabilities are based on league-wide results and do not take into account the properties of specific teams, let us call the teams Team A and Team B.

Before the game starts, the probability that Team A wins is $0.50$.

Now, suppose that Team A receives the opening kickoff, which results
in a touchback so that Team A starts with the ball at their own 25 -yard
line. A touchback in football is when the ball becomes dead on or behind the goal line a team is defending, provided that the impetus comes from an opponent and that it is not a touchdown or an incomplete pass

This additional information changes the probability that Team A wins.

Using the win probability calculator on the Pro Football Reference website
(attps://ww. pro-football-reference. com), that probability is now $0.52$.

This may be viewed as the conditional probability that Team A wins the game
given that the team received the opening kickoff and it is a touchback.


As the game unfolds, the probability that Team A wins the game evolves,
depending on the specific outcomes.

For instance, if on the first play from scrimmage, Team A has a 30 -yard pass completion, taking 10 seconds, then
their win probability increases from $0.52$ to $0.58$.

If on the next play, the quarterback is sacked, fumbles, and loses the ball at his own 45 , the probability
that Team A wins drops to $0.42$.

These win probabilities can be viewed as the conditional probabilities that a team wins the game given the current status (pasibly hypothetical) of the game.

You can read a bit more about winning probabilities in [this blog post](https://www.the33rdteam.com/win-probability-explained/). These computations can be done also via simulation (see the NYT article from before)
